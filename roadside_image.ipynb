```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/road_side1.jpg")
img
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.subplot(1, 1, 1)
plt.imshow(img_rgb)
plt.show()
```

```python
xml_data = cv2.CascadeClassifier('xml_data.xml')
  
found = xml_data.detectMultiScale(img_gray, 
                                   minSize =(20, 20))
```

```python
amount_found = len(found)
```

```python
if amount_found != 0:
  for (x, y, width, height) in found:
    cv2.rectangle(img_rgb, (x, y), 
                      (x + height, y + width), 
                      (0, 255, 0), 5)
```

```python
plt.subplot(1, 1, 1)
plt.imshow(img_rgb)
plt.show()
```
